**World happiness index**

```python
import tensorflow as tf 

import pandas as pd

import matplotlib.pyplot as plt

happy_index=pd.read_csv("world-happiness-report-2021.csv")

happy_index
```

```python
happy_index["Country name"]
```

```python
index_one_hot=pd.get_dummies(happy_index)
index_one_hot.head()
```

```python
x=index_one_hot.drop("Ladder score",axis=1)
y=index_one_hot["Ladder score"]  #Ladder score
```

```python
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=30)

X_train # x values under training 
```

```python
y_train
```

```python
len(x),len(X_train),len(X_test)
```

```python
0.3*149
```

```python
model=tf.keras.Sequential([tf.keras.layers.Dense(10),tf.keras.layers.Dense(1)])
model.compile(loss=tf.keras.losses.mae,optimizer=tf.keras.optimizers.SGD(),metrics=['mae'])
model.fit(X_train,y_train,epochs=100)
```

```python
model.evaluate(X_test,y_test)
```
